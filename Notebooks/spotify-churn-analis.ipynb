# Data Loading

```python
df = pd.read_csv('dataset_spotify_cleaned.csv')
```

# Data Preparation

```python
# ubah semua kolom bertipe object/category jadi lowercase
for col in df.select_dtypes(include=['object', 'category']).columns:
    df[col] = df[col].str.lower()

df.head(10)
```
